```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load a grayscale image
image = cv2.imread('sample.jpg', 0)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original and equalized images
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(equalized_image, cmap='gray'), plt.title('Equalized Image')
plt.show()
```

```python
//ERSOION AND DILATION
import cv2
import numpy as np

# Load the image
img = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)

# Define the structuring element (kernel)
kernel = np.ones((3, 3), np.uint8)

# Apply erosion
eroded_img = cv2.erode(img, kernel, iterations=1)

# Apply dilation
dilated_img = cv2.dilate(img, kernel, iterations=1)

# Display the original, eroded, and dilated images
cv2.imshow('Original', img)
cv2.imshow('Eroded', eroded_img)
cv2.imshow('Dilated', dilated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
//OPENING CLOSING
import cv2
import numpy as np

# Load a binary image
image = cv2.imread('samp2.jpg', 0)

# Check if the image was loaded correctly
if image is None:
    print("Error: Image not found or unable to load.")
else:
    # Define a structuring element
    kernel = np.ones((5,5), np.uint8)

    # Apply opening
    opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)

    # Apply closing
    closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

    # Display the results
    cv2.imshow('Opening', opening)
    cv2.imshow('Closing', closing)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
    //EDGE DETECTION
import cv2

# Load the image
img = cv2.imread('samp2.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Sobel edge detection
grad_x = cv2.Sobel(img, cv2.CV_8U, 1, 0, ksize=3)
grad_y = cv2.Sobel(img, cv2.CV_8U, 0, 1, ksize=3)
grad = cv2.addWeighted(grad_x, 0.5, grad_y, 0.5, 0)

# Apply Canny edge detection
canny_edges = cv2.Canny(img, 50, 150)

# Display the original, Sobel, and Canny edge detection results
cv2.imshow('Original', img)
cv2.imshow('Sobel Edges', grad)
cv2.imshow('Canny Edges', canny_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
// IMAGE DISTORTION
import cv2
import numpy as np

# Load the image
img = cv2.imread('sample.jpg')

# Define the distortion coefficients
dist_coeffs = np.array([-0.1, 0.2, 0, 0, 0])

# Define the camera matrix
camera_matrix = np.array([[1000, 0, 320], [0, 1000, 240], [0, 0, 1]])

# Apply image warping
warp_img = cv2.undistort(img, camera_matrix, dist_coeffs)

# Display the original and warped images
cv2.imshow('Original', img)
cv2.imshow('Warped', warp_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
// IMAGE SmOOTHING
import cv2
import numpy as np

# Load the image
img = cv2.imread('image.jpg')

# Apply Gaussian Blur
blurred_img = cv2.GaussianBlur(img, (5, 5), 0)

# Apply Median Blur
median_blurred_img = cv2.medianBlur(img, 5)

# Apply Bilateral Filtering
bilateral_blurred_img = cv2.bilateralFilter(img, 5, 50, 50)

# Display the original and smoothed images
cv2.imshow('Original', img)
cv2.imshow('Gaussian Blur', blurred_img)
cv2.imshow('Median Blur', median_blurred_img)
cv2.imshow('Bilateral Filtering', bilateral_blurred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
